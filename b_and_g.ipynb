```
import pandas as pd
import gzip
```

```
df = pd.read_json("http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz" , lines=True)
df.head()
```

```
vowels = ['a','e','i','o','u','A','E','I','O','U']
```

```
import re
```

```
def splitstrings(str):
  str = re.sub('[^a-zA-Z]',' ', str)
  return list(str)
```

```
def letter(text):
  dict = {"vowels":0,
          "consonants":0}
  for word in text:
    words = splitstrings(word)
    for x in words:
      if x in vowels:
        dict["vowels"] += 1
      else:
         dict["consonants"] += 1
  return dict
```

```
df['reviewText_vccounts'] = df['reviewText'].apply(letter)
df['summary_vccounts'] = df['summary'].apply(letter)
df.head()
```

```
def wordsbeginb(string):
    string = re.sub('[^a-zA-Z]',' ', string) 
    list_1 = re.findall(r'\b[bB]\w+', string)

    return(list_1)
```

```
df["reviewText_wordsbeginb"] = df.reviewText.apply(wordsbeginb)
df["summary_wordsbeginb"] = df.summary.apply(wordsbeginb)
df.head()
```

```
def wordsendg(string):
    string = re.sub('[^a-zA-Z]',' ', string) 
    list_2= re.findall(r'\w+[gG]\b', string) 

    return(list_2)
```

```
df["reviewText_wordsendg"] = df.reviewText.apply(wordsendg)
df["summary_wordsendg"] = df.summary.apply(wordsendg)
df.head()
```

```
def wordsbeginendbgboth(string):
    string = re.sub('[^a-zA-Z]',' ', string) 
    list_3= re.findall(r'\b[bB]\w+[gG]\b', string) 

    return(list_3)
    
df["reviewText_BG"] = df.reviewText.apply(wordsbeginendbgboth)
df["summary_BG"] = df.summary.apply(wordsbeginendbgboth)
df.head()
```

```
def unionbg(string):
    string= re.sub('[^a-zA-Z]',' ', string) 
    list4 = re.findall(r'\b[bB]\w+', string) 
    list5 = re.findall(r'\w+[gG]\b', string) 
    return (list(set().union(list4, list5)))

df["reviewText_union"] = df.reviewText.apply(unionbg)
df["summary_union"] = df.summary.apply(unionbg)
df.head()
```
